```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è id –∑–∞–ø–∏—Å–µ–π —Å–æ–æ–±—â–µ—Å—Ç–≤–∞
def get_posts(group_id, offset, count):
  posts = vk_api.wall.get(owner_id=-group_id, offset=offset, count=count, extended=0, v=5.131)['items']
  return posts
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ –∑–∞–ø–∏—Å–µ–π —Å–æ–æ–±—â–µ—Å—Ç–≤–∞
def get_posts_count(group_id):
  return vk_api.wall.get(owner_id=-group_id, offset=0, count=1, extended=0, v=5.131)['count']
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è id –∏ —Ç–µ–∫—Å—Ç–∞ –∑–∞–ø–∏—Å–µ–π —Å–æ–æ–±—â–µ—Å—Ç–≤–∞, count x 100
def get_posts_n(group_id, count):

  for i in range(count):
    posts = get_posts(group_id, i * 100, 100)

  posts_id = [post['id'] if 'id' in post else None for post in posts]
  posts_owner_id = [post['owner_id'] if 'owner_id' in post else None for post in posts]
  posts_from_id = [post['from_id'] if 'from_id' in post else None for post in posts]
  posts_created_by = [post['created_by'] if 'created_by' in post else None for post in posts]
  posts_date = [post['date'] if 'date' in post else None for post in posts]
  posts_text = [post['text'] if 'text' in post else None for post in posts]
  posts_reply_post_id = [post['reply_post_id'] if 'reply_post_id' in post else None for post in posts]
  posts_reply_owner_id = [post['reply_owner_id'] if 'reply_owner_id' in post else None for post in posts]
  posts_copyright = [post['copyright'] if 'copyright' in post else None for post in posts]
  posts_likes = [post['likes'] if 'likes' in post else None for post in posts]
  posts_reposts = [post['reposts'] if 'reposts' in post else None for post in posts]
  posts_post_type = [post['post_type'] if 'post_type' in post else None for post in posts]
  posts_attachments = [post['attachments'] if 'attachments' in post else None for post in posts]
  posts_copy_history = [post['copy_history'] if 'copy_history' in post else None for post in posts]
  posts_comments = [post['comments'] if 'comments' in post else None for post in posts]


  return pd.DataFrame(data={'id': posts_id,
                            'owner_id': posts_owner_id,
                            'from_id': posts_from_id,
                            'created_by': posts_created_by,
                            'date': posts_date,
                            'text': posts_text,
                            'reply_post_id': posts_reply_post_id,
                            'reply_owner_id': posts_reply_owner_id,
                            'copyright': posts_copyright,
                            'likes': posts_likes,
                            'reposts': posts_reposts,
                            'post_type': posts_post_type,
                            'attachments': posts_attachments,
                            'copy_history': posts_copy_history,
                            'comments': posts_comments
                            })
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –≤—Å–µ—Ö id –∏ —Ç–µ–∫—Å—Ç–∞ –∑–∞–ø–∏—Å–µ–π —Å–æ–æ–±—â–µ—Å—Ç–≤–∞
def get_posts_all(group_id):

  count = get_posts_count(group_id)

  iters = count // 100
  rest = count % 100

  posts = []

  for i in tqdm(range(iters)):
    posts.extend(get_posts(group_id, i * 100, 100))

  posts.extend(get_posts(group_id, iters * 100, rest))


  posts_id = [post['id'] if 'id' in post else None for post in posts]
  posts_owner_id = [post['owner_id'] if 'owner_id' in post else None for post in posts]
  posts_from_id = [post['from_id'] if 'from_id' in post else None for post in posts]
  posts_created_by = [post['created_by'] if 'created_by' in post else None for post in posts]
  posts_date = [post['date'] if 'date' in post else None for post in posts]
  posts_text = [post['text'] if 'text' in post else None for post in posts]
  posts_reply_post_id = [post['reply_post_id'] if 'reply_post_id' in post else None for post in posts]
  posts_reply_owner_id = [post['reply_owner_id'] if 'reply_owner_id' in post else None for post in posts]
  posts_copyright = [post['copyright'] if 'copyright' in post else None for post in posts]
  posts_likes = [post['likes'] if 'likes' in post else None for post in posts]
  posts_reposts = [post['reposts'] if 'reposts' in post else None for post in posts]
  posts_post_type = [post['post_type'] if 'post_type' in post else None for post in posts]
  posts_attachments = [post['attachments'] if 'attachments' in post else None for post in posts]
  posts_copy_history = [post['copy_history'] if 'copy_history' in post else None for post in posts]
  posts_comments = [post['comments'] if 'comments' in post else None for post in posts]


  return pd.DataFrame(data={'id': posts_id,
                            'owner_id': posts_owner_id,
                            'from_id': posts_from_id,
                            'created_by': posts_created_by,
                            'date': posts_date,
                            'text': posts_text,
                            'reply_post_id': posts_reply_post_id,
                            'reply_owner_id': posts_reply_owner_id,
                            'copyright': posts_copyright,
                            'likes': posts_likes,
                            'reposts': posts_reposts,
                            'post_type': posts_post_type,
                            'attachments': posts_attachments,
                            'copy_history': posts_copy_history,
                            'comments': posts_comments
                            })
```

##Comments

```python
import itertools
  
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞–Ω–Ω—ã—Ö –≤—Å–µ—Ö –∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–µ–≤ —Å–ø–∏—Å–∫–∞ –∑–∞–ø–∏—Å–µ–π —Å–æ–æ–±—â–µ—Å—Ç–≤–∞
def get_posts_comments_all(group_id, posts_id):
  
  obs, cbs = "{","}"

  comments = []
  
  post_i = 0

  posts_len = len(posts_id)
  pbar = tqdm(total=posts_len)

  softlock_counter = 0

  while post_i < posts_len:
    try:
      response = vk_api.execute(code = f"\
      var R, Res = [],\
      group_id = {group_id},\
      post_ids = {posts_id[post_i:post_i+26]},\
      id = 0,\
      offset = 0,\
      count = 100, loop=0;\
      \
      while(loop < 25) {obs}\
        R = API.wall.getComments({obs} \
          \"owner_id\": -group_id,\
          \"post_id\": post_ids[id],\
          \"offset\": offset,\
          \"count\": count,\
          \"extended\": 0,\
          \"thread_items_count\": 10{cbs});\
      var len = R.items.length;\
      if (len > 0) \
        Res.push(R);\
      if (len == 100) \
        offset = offset + 100;\
      if (len < 100) {obs}\
        offset = 0;\
        id = id + 1; {cbs}\
      loop = loop + 1;\
      {cbs}\
      return [id, Res];", v=5.131)

      it_delta = max(1,response[0])
      post_i += it_delta

      softlock_counter = 0

      newcomms = list(itertools.chain( *[r['items'] for r in response[1]] ))

      comments.extend(newcomms)
      pbar.update(it_delta)
      time.sleep(0.35)

    except Exception as ex:
      print(ex)

      if (softlock_counter > 10):
        post_i = posts_len

      softlock_counter += 1
      time.sleep(0.5)




  l2_comments = []
  for comment in comments:
    if comment['thread']['count'] > 0:
      l2_comments.extend(comment['thread']['items'])

  comments.extend(l2_comments)

  comments_date = []
  comments_from_id = []
  comments_id = []
  comments_owner_id = []
  comments_post_id = []
  comments_text = []

  for i in range(len(comments)):
    if comments[i]['from_id'] != 0: #filter deleted
      comments_date.append(comments[i]['date'] if 'date' in comments[i] else None)
      comments_from_id.append(comments[i]['from_id'] if 'from_id' in comments[i] else None)
      comments_id.append(comments[i]['id'] if 'id' in comments[i] else None)
      comments_owner_id.append(comments[i]['owner_id'] if 'owner_id' in comments[i] else None)
      comments_post_id.append(comments[i]['post_id'] if 'post_id' in comments[i] else None)
      comments_text.append(comments[i]['text'] if 'text' in comments[i] else None)
  
  pbar.close()

  return pd.DataFrame(data={
      'date': comments_date,
      'from_id': comments_from_id,
      'id': comments_id,
      'owner_id': comments_owner_id,
      'post_id': comments_post_id,
      'text': comments_text})
```

##User Friends and Subscriptions

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ –¥—Ä—É–∑–µ–π –∏ –ø–æ–¥–ø–∏—Å–æ–∫ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
# –≤ —Å–ª—É—á–∞–µ –ø—Ä–∏–≤–∞—Ç–Ω–æ—Å—Ç–∏ - –ø–æ–¥–Ω—è—Ç —Ñ–ª–∞–≥ private
def get_user_friends_subscriptions(user_id):
  #check privacy
  try:
    prv_test = vk_api.friends.get(user_id=user_id, offset = 0,
                                  count=1, v=5.131)
    count = prv_test['count']


    #FRIENDS
    
    user_friends = vk_api.friends.get(user_id=user_id,offset = 0,
                                      count=count, v=5.131)['items']

    #SUBSCRIPTIONS

    subscr_count = vk_api.users.getSubscriptions(user_id=user_id, offset = 0,
                                                 count=1, v=5.131)['groups']
    count = subscr_count['count']

    iters = count // 200
    rest = count % 200

    user_subscriptions = []

    for i in range(iters):
      subscriptions = vk_api.users.getSubscriptions(user_id=user_id, offset = i * 200,
                                                 count=200, v=5.131)['groups']['items']
      user_subscriptions.extend(subscriptions)

    subscriptions = vk_api.users.getSubscriptions(user_id=user_id, offset = iters * 200,
                                                 count=rest, v=5.131)['groups']['items']
    user_subscriptions.extend(subscriptions)

    #RESULT

    return {'user_id': user_id, 'private': 0,
            'friends': user_friends, 'subscriptions': user_subscriptions}

  except vk.exceptions.VkAPIError as ex:
    return {'user_id': user_id, 'private': 1, 'friends': [], 'subscriptions': []}
```

```python
# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å–ø–∏—Å–∫–∞ –¥—Ä—É–∑–µ–π –∏ –ø–æ–¥–ø–∏—Å–æ–∫ —Å–ø–∏—Å–∫–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π
# –≤ —Å–ª—É—á–∞–µ –ø—Ä–∏–≤–∞—Ç–Ω–æ—Å—Ç–∏ - –ø–æ–¥–Ω—è—Ç —Ñ–ª–∞–≥ private
def get_users_friends_subscriptions(user_ids):

  user_ids = list(set(user_id for user_id in user_ids if user_id > 0))
  obs, cbs = "{","}"
  user_names_data = []
  user_id = []
  user_private = []
  user_friends = []
  user_subscriptions = []

  softlock_counter = 0

  req_fields = ["blacklisted",
                "country",
                "education",
                "followers_count",
                "has_photo",
                "home_town",
                "interests",
                "has_mobile",
                "occupation",
                "religion,inspired_by",
                "langs, political",
                "verified, career",
                "deactivated",
                "is_closed"]
                
  users_len = len(user_ids)
  user_i = 0
  
  
  pbar0 = tqdm(total=users_len)
      

  while user_i < users_len:
    try:
      user_upper = min(users_len, user_i+1000)
      response = vk_api.execute(code = f"\
      var R, Res = [],\
      fields = {req_fields},\
      user_ids = {user_ids[user_i:user_upper]},\
      count = 40, loop=0;\
      \
      var arrlen = user_ids.length;\
      while(loop < 25 && count * loop < arrlen) {obs}\
        R = API.users.get({obs} \
          \"user_ids\": user_ids.slice(count * loop, count * (1 + loop)),\
          \"fields\": fields{cbs});\
        var len = R.length;\
        Res.push(R);\
        if (len < 40)\
          loop = 25;\
        loop = loop + 1;\
      {cbs}\
      return Res;", v=5.131)
      softlock_counter = 0

      newdata = list(itertools.chain( *response ))
      ndlen = len(newdata)

      
      pbar0.update(ndlen)

      user_i += ndlen

      user_names_data.extend(newdata)
      time.sleep(0.35)


    except Exception as ex:
      print(ex)

      if (softlock_counter > 10):
        user_i = users_len

      softlock_counter += 1
      time.sleep(0.5)

  pbar0.close()


  user_first_name = [user['first_name'] for user in user_names_data]
  user_last_name = [user['last_name'] for user in user_names_data]

  user_blacklisted = [user['blacklisted'] if 'blacklisted' in user else None for user in user_names_data]
  user_country = [user['country'] if 'country' in user else None for user in user_names_data]
  user_education = [user['education'] if 'education' in user else None for user in user_names_data]
  user_followers_count = [user['followers_count'] if 'followers_count' in user else None for user in user_names_data]
  user_has_photo = [user['has_photo'] if 'has_photo' in user else None for user in user_names_data]
  user_home_town = [user['home_town'] if 'home_town' in user else None for user in user_names_data]
  user_interests = [user['interests'] if 'interests' in user else None for user in user_names_data]
  user_has_mobile = [user['has_mobile'] if 'has_mobile' in user else None for user in user_names_data]
  user_occupation = [user['occupation'] if 'occupation' in user else None for user in user_names_data]
  user_religion = [user['religion'] if 'religion' in user else None for user in user_names_data]
  user_inspired_by = [user['inspired_by'] if 'inspired_by' in user else None for user in user_names_data]
  user_langs = [user['langs'] if 'langs' in user else None for user in user_names_data]
  user_political = [user['political'] if 'political' in user else None for user in user_names_data]
  user_verified = [user['verified'] if 'verified' in user else None for user in user_names_data]
  user_career = [user['career'] if 'career' in user else None for user in user_names_data]
  user_deactivated = [user['deactivated'] if 'deactivated' in user else None for user in user_names_data]
  user_is_closed = [user['is_closed'] if 'is_closed' in user else None for user in user_names_data]

  ufl_i = 0
  ufl_private = []
  ufl_friends = []
  ufl_subscriptions = []

  pbar = tqdm(total=users_len)

  while ufl_i < users_len:
    post_upper = min(users_len, ufl_i+12)
    try:
      response = vk_api.execute(code = f"\
      var R1, R2, Res1 = [], Res2 = [], Res3 = [],\
      user_ids = {user_ids[ufl_i:post_upper]},\
      id = 0,\
      offset = 0,\
      prv = 1,\
      npr = 0,\
      usrlen = user_ids.length,\
      loop=0;\
      while(loop < 12 && id < usrlen) {obs}\
        R1 = API.users.getSubscriptions({obs} \
          \"user_id\": user_ids[id],\
          \"count\": 200,\
          \"offset\": 0{cbs});\
        R2 = API.friends.get({obs} \
          \"user_id\": user_ids[id],\
          \"count\": 1000,\
          \"offset\": 0{cbs});\
        var len = R2.items.length;\
        Res1.push(R1);\
        Res2.push(R2);\
        if (len == 0){obs}\
          Res3.push(prv);{cbs}\
        else{obs}\
          Res3.push(npr);{cbs}\
        id = id + 1;\
        loop = loop + 1;\
        {cbs}\
      return [Res1, Res2, Res3];", v=5.131)

      it_delta = len(response[0])
      ufl_i += it_delta

      softlock_counter = 0

      ufl_private.extend(response[2])
      ufl_friends.extend(response[1])
      ufl_subscriptions.extend(response[0])

      pbar.update(it_delta)
      time.sleep(0.35)

    except Exception as ex:
      print(ex)

      softlock_counter += 1
      time.sleep(0.5)

  user_private = ufl_private
  user_friends = [x['items'] if type(x) != bool and 'items' in x else None for x in ufl_friends]
  user_subscriptions = [x['groups']['items']  
                        if type(x) != bool
                        and 'groups' in x
                        and 'items' in x['groups']
                        else None for x in ufl_subscriptions]
  pbar.close()
  
  return pd.DataFrame(data={'id': user_ids,
                            'private': user_private,
                            'friends': user_friends,
                            'subscriptions': user_subscriptions,
                            'first_name': user_first_name,
                            'last_name': user_last_name,
                            'blacklisted': user_blacklisted,
                            'country': user_country,
                            'education': user_education,
                            'followers_count': user_followers_count,
                            'has_photo': user_has_photo,
                            'home_town': user_home_town,
                            'interests': user_interests,
                            'has_mobile': user_has_mobile,
                            'occupation': user_occupation,
                            'religion': user_religion,
                            'inspired_by': user_inspired_by,
                            'langs': user_langs,
                            'political': user_political,
                            'verified': user_verified,
                            'career': user_career,
                            'deactivated': user_deactivated,
                            'is_closed': user_is_closed
                            })
```
